```python
# Tkinter,
import tkinter as tk
# widh => Button
class BasicCalculator():
    def __init__(self,root):
        self.root=root
        self.root.title("Basic Calculator")
        self.root.geometry("300x400")
        self.root.resizable(False,False)

        self.expression = ""
        self.entry = tk.Entry(root,font=("Arial",20), bd=10, insertborderwidth=2, width=14, border=4, justify="right")
        self.entry.grid(row=0,column=0, columnspan=4)
        self.create_buttons()
    
    def create_buttons(self):
        button = [
            ("7",1,0),("8",1,1),("9",1,2),("/",1,3),
            ("4",2,0),("5",2,1),("6",2,2),("*",2,3),
            ("1",3,0),("2",3,1),("3",3,2),("-",3,3),
            ("0",4,0),(".",4,1),("=",4,2),("+",4,3),
            ("C",5,0)
            ]
        for (text,row,col) in button:
            if text == "=":
                btn = tk.Button(self.root,text=text,padx=20,pady=20,font=("Arial",14),bg="lightblue", command=self.calculate)
            elif text == "C":
                btn = tk.Button(self.root,text=text,padx=132,pady=20,font=("Arial",14),bg="orange", command = self.clear)
                btn.grid(row=row, column=col, columnspan=4)
                continue
            else:
                btn = tk.Button(self.root,text=text,padx=21,pady=19,font=("Arial",14), command=lambda t=text: self.press(t))
            btn.grid(row=row, column=col)

    def press(self,char):
        self.expression += char
        self.entry.delete(0,tk.END)
        self.entry.insert(tk.END,self.expression)
    
    def clear(self):
        self.expression =""
        self.entry.delete(0,tk.END)
   
    def calculate(self):
        try:
            result= str(eval(self.expression))
            self.entry.delete(0,tk.END)
            self.entry.insert(tk.END,result)
            print(result)
        except:
            self.entry.delete(0,tk.END)
            self.entry.insert(tk.END,"ERROR")
            self.expression =""
        
root = tk.Tk()  # Window object
app = BasicCalculator(root)
root.mainloop()  # Ubdating UI
```
